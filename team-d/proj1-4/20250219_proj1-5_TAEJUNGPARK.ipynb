```python
import pandas as pd

# Load the uploaded CSV file for EDA
file_path = "/mnt/data/ad_data (1).csv"
df = pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert date columns to datetime for analysis
df['광고_시작일'] = pd.to_datetime(df['광고_시작일'])
df['광고_종료일'] = pd.to_datetime(df['광고_종료일'])

# Add a derived column: 광고_기간 (duration of the ad)
df['광고_기간'] = (df['광고_종료일'] - df['광고_시작일']).dt.days

# Calculate correlation matrix to find strong relationships (potential AHA moments)
correlation_matrix = df.corr()

# Plot correlation heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap', fontsize=16)
plt.show()

# Extract top correlations with 매출액 to identify potential AHA moments
top_corr_with_sales = correlation_matrix['매출액'].sort_values(ascending=False)
top_corr_with_sales
```
